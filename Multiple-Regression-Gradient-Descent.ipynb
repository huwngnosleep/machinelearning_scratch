```python
data["sqft_living"] = data["sqft_living"] / 1000
features = data[['constant', 'bedrooms', 'bathrooms', 'sqft_living']]
output = data['price'] / 100000
features, output
```

```python
def predict_outcome(feature_matrix, weights):
    predictions = np.dot(feature_matrix, weights)
    return(predictions)
```

```python
def feature_derivative(errors, feature):
    derivative = 2 * np.dot(errors, feature)
    return(derivative)
```

```python
feature_name = ['constant', 'bedrooms', 'bathrooms', 'sqft_living']
def regression_gradient_descent(feature_matrix, output, initial_weights, step_size, tolerance):
    converged = False
    weights = np.array(initial_weights)
    while not converged:
        # compute the predictions based on feature_matrix and weights:
        predictions = predict_outcome(feature_matrix, weights)
        # compute the errors as predictions - output:
        errors = np.absolute(predictions - output)
        gradient_sum_squares = 0 # initialize the gradient
        # while not converged, update each weight individually:
        for i in range(len(weights)):
            # Recall that feature_matrix[:, i] is the feature column associated with weights[i]
            # compute the derivative for weight[i]:
            derivative = feature_derivative(errors, feature_matrix[feature_name[i]])
            # add the squared derivative to the gradient magnitude
            gradient_sum_squares += np.square(derivative)
            # update the weight based on step size and derivative:
            weights[i] = step_size * derivative
            print(i, feature_name[i], weights[i], derivative)
            print(weights)
        gradient_magnitude = math.sqrt(gradient_sum_squares)
        if gradient_magnitude < tolerance:
            converged = True
    return(weights)
```

```python
weights = regression_gradient_descent(features, output, [0, 0], 0.05, 0.01)
print(weights)
```
