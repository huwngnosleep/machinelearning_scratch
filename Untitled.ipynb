```python
import mammoth

with open('test-vat-lt.docx', "rb") as docx_file:
    result = mammoth.convert_to_html(docx_file)
    text = result.value # The raw text
    print(text)
```

```python
import docx
doc = docx.Document('test-vat-lt.docx')  # Creating word reader object.
data = ""
fullText = []
for para in doc.paragraphs:
    fullText.append(para.text)
    data = '\n'.join(fullText)

print(data)
```
